```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
stock_portfolio=['AMZN','GOOGL','MSFT']
stock_data=yf.download(stock_portfolio,start='2024-06-01',end='2024-11-30')['Adj Close']
print(stock_data.head())
print(stock_data.tail())
```

```python
weights=[1/3,1/3,1/3]
portfolio_daily_ret=stock_data.pct_change()[1:]
print(portfolio_daily_ret.head())
print(portfolio_daily_ret.tail())
```

```python
portfolio_weighted_daily_ret=(weights*portfolio_daily_ret)
portfolio_return=portfolio_weighted_daily_ret.sum(axis=1)
invested_amount=10000
portfolio_value=invested_amount*(1+portfolio_return).cumprod()
print(portfolio_value)
```

```python
plt.plot(portfolio_value,color='#88c999')
plt.title('Portfolio Value over the last 6 months',color='r')
plt.xlabel('Dates',color='hotpink')
plt.ylabel('Portfolio Value',color='hotpink')
```

```python
portfolio_final_value=portfolio_value.iloc[-1]
print('The total value of the portfolio at the end of last 6 months is $'+str(portfolio_final_value))
```
