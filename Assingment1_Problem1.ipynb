```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
stock_data=yf.download('AAPL',start='2023-01-01',end='2023-12-31')
print(stock_data.head())
print(stock_data.tail())
dates=np.array(stock_data.index)
closing_prices=np.array(stock_data['Adj Close'])
```

```python
plt.plot(dates,closing_prices,label='Apple closing prices',lw='2',color='c')
plt.xlabel('Date',color='hotpink')
plt.ylabel('Closing Price',color='hotpink')
plt.title('APPLE stock closing prices 2023')
```

```python
print('Based on the plot given above of AAPL closing stock price the general trend is upward.')
print('The stock price showed some fluctuations throughout the year but overall the trend was upward,as the stock price closing price was $123.76 on 2023-1-3 and which increased to $191.59 on 2023-12-29.')
print('This shows a general upward trend of AAPL stock overthe year 2023.')
```
